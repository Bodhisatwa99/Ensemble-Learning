```python
#  43
breast_cancer = load_breast_cancer()
X_bc, y_bc = breast_cancer.data, breast_cancer.target
X_train_bc, X_test_bc, y_train_bc, y_test_bc = train_test_split(X_bc, y_bc, test_size=0.3, random_state=random_state)
rf_pr_clf = RandomForestClassifier(n_estimators=100, random_state=random_state)
rf_pr_clf.fit(X_train_bc, y_train_bc)
y_prob_rf_pr = rf_pr_clf.predict_proba(X_test_bc)[:, 1]
precision, recall, _ = precision_recall_curve(y_test_bc, y_prob_rf_pr)
pr_auc = auc(recall, precision)
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='b', lw=2, label=f'Precision-Recall curve (AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('43. Random Forest Classifier Precision-Recall Curve')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```

```python
#44. 
X_stack, y_stack = make_classification(n_samples=100, n_features=10, random_state=random_state)
X_train_stack, X_test_stack, y_train_stack, y_test_stack = train_test_split(X_stack, y_stack, test_size=0.3, random_state=random_state)
estimators_stack = [
    ('rf', RandomForestClassifier(n_estimators=50, random_state=random_state)),
    ('lr', LogisticRegression(random_state=random_state, solver='liblinear'))
]
stacking_rf_lr = StackingClassifier(estimators=estimators_stack, final_estimator=LogisticRegression(random_state=random_state, solver='liblinear'), cv=2)
stacking_rf_lr.fit(X_train_stack, y_train_stack)
accuracy_stacking_rf_lr = accuracy_score(y_test_stack, stacking_rf_lr.predict(X_test_stack))
print("44. Stacking Classifier (RF, LR) Accuracy:", accuracy_stacking_rf_lr)

rf_clf_single = RandomForestClassifier(n_estimators=100, random_state=random_state)
rf_clf_single.fit(X_train_stack, y_train_stack)
accuracy_rf_single = accuracy_score(y_test_stack, rf_clf_single.predict(X_test_stack))
print("44. Single Random Forest Accuracy:", accuracy_rf_single)

lr_clf_single = LogisticRegression(random_state=random_state, solver='liblinear')
lr_clf_single.fit(X_train_stack, y_train_stack)
accuracy_lr_single = accuracy_score(y_test_stack, lr_clf_single.predict(X_test_stack))
print("44. Single Logistic Regression Accuracy:", accuracy_lr_single)

print("44. Comparison: Stacking Accuracy - {:.4f}, RF Accuracy - {:.4f}, LR Accuracy - {:.4f}".format(
    accuracy_stacking_rf_lr, accuracy_rf_single, accuracy_lr_single
))
```
